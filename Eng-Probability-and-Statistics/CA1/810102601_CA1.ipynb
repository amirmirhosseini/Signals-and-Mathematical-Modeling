# ۱-۲ )
تعریف تابع برای محاسبه مثدار تئوری واریناس و میانگین

```r
theomean <- function(mm , NN , KK){
    meannn <- mm*(KK/NN)
    return(meannn)
}  
theovar <- function(mm  , NN , KK){
    varrr <- mm*KK*(NN-KK)*(NN-mm)/(NN*NN*(NN-1))
    return(varrr)
}  
```

تعریف تابع برای محاسبه مقدار عملی توزیع، خروجی این تابع میانگین، واریانس و تعداد دفعات شبیه‌سازی می‌باشد.

```r

q1p2 <- function(N0 , m0 , k0){
    Q <- ((10000-100)/50)
    s = 100
    sim_num <- numeric(Q)
    resultsmean <- numeric(Q)
    resultsvar <- numeric(Q)
    for(i in 1:Q){
        presults <- rhyper(s,k0,N0-k0,m0)
        resultsmean[i] <- mean(presults)
        resultsvar[i] <- var(presults)
        sim_num[i] <- s
        
        s <- (s+50)
    }
    return(list(meanbro = resultsmean, varbro = resultsvar , snumbro = sim_num))  
}
```

* پیاده‌‌سازی توابع

* 
به ترتیب توزیع میانگین ، توزیع واریانس ، میانگین توزیع میانگین ، میانگین توزیع واریانس ، مقدار تئوری واریانس ، مقدار تئوری میانگین

```r
bro <- q1p2(N,m,k)
cat(bro$meanbro , '\n\n\n')
cat(bro$varbro , '\n\n\n')
cat(mean(bro$meanbro),'\n')
cat(mean(bro$varbro),'\n')
theorvar = theovar(m,N,k)
theormean = theomean(m,N,k)
cat(theorvar,'\n')
cat(theormean)
```

```r
expoDist <- function(ed) {
    edminus <- c()
    
    for (i in 1:length(ed)) {
        if (ed[i] > 12) {
            edminus <- c(edminus, ed[i] - 12)
        }
    }
    return(edminus)  
}
```

# ۳-۳)

```r

mval <- c(1000, 10000, 100000)
edi <- list() 
ediminus <- list() 

for (j in 1:3) {
    a <- expDist(1/15, mval[j])
    edi[[j]] <- a 
}

for (k in 1:3) {
    a <- expoDist(edi[[k]])
    ediminus[[k]] <- a  
}

# Plotting
par(mfrow = c(2, 3)) 
for (i in 1:3) {
  
    hist(edi[[i]],
         breaks = seq(floor(min(edi[[i]])) - 0.5, ceiling(max(edi[[i]])) + 0.5, by = 1),
         main = paste("m =", mval[i]),  
         xlab = "minutes",
         ylab = "Frequency",
         col = "lightyellow",
         xlim = c(floor(min(edi[[i]])), ceiling(max(edi[[i]]))) 
    )
    
    if (length(ediminus[[i]]) > 0) {
        hist(ediminus[[i]],
             breaks = seq(floor(min(ediminus[[i]])) - 0.5, ceiling(max(ediminus[[i]])) + 0.5, by = 1),
             main = paste("m =", mval[i], "after 12 minutes"),
             xlab = "minutes",
             ylab = "Frequency",
             col = "darkblue",
             xlim = c(floor(min(ediminus[[i]])), ceiling(max(ediminus[[i]])))  
        )
    } else {
        plot.new() 
        title(main = paste("m =", mval[i], "after 12 minutes"), col.main = "darkblue")
    }
}
```
